```python
import pandas as pd

Location = 'titanic3.xls'
print(Location)
```

```python
df = pd.read_excel(Location)
```

```python
df.describe()
```

```python
df1 = df.drop(['name','ticket','cabin','boat','body','home.dest'], axis=1)
```

```python
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
df1['sex_val'] = df1['sex'].apply(score_to_numeric)
df1.tail()
```

```python
df1['embarked'].unique()
```

```python
df_embarked = pd.get_dummies(df1['embarked'])
df_embarked.tail()
```

```python
df_new2 = pd.concat([df1, df_embarked], axis=1)
df_new2.tail()
```

```python
df_new2["age"].fillna(df.groupby("sex")["age"].transform("mean"), inplace=True)
df_new2.describe()
```

```python
df_new3 = df_new2.drop(['sex'], axis=1)
df_new3.head()
```

```python
df_new3.corr()
```

```python
#the lower passenger's class - the lower chances for survival
#the higher fare paid - the higher chances for survival
#rates of survival higher for people embarked at C = Cherbourg
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ pclass + age + sibsp + parch + fare + sex_val + C + Q + S -1', data=df_new3).fit()
result.summary()
```

```python
# for parch and fare P is higher then 0.5 so we have to exclude them from equation
```

```python
result = sm.ols(formula='survived ~ pclass + age + sibsp + sex_val + C + Q + S -1', data=df_new3).fit()
result.summary()
```
