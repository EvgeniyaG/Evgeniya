```python
#Axis Auto Sales
#Carlos Hugens is the Sales Manager for Axis Auto Sales a low cost regional chain of Used Car Lots.
#Carlos is getting ready for his annual sales meeting and he is looking for the best way to improve his
#sales group's performance.
#His data includes the gender, years of experience, sales training and hours worked per week. It also
#includes the average cars sold per month by each salesperson.
#Find out the following:
#a) average cars sold per month
#b) max cars sold per month
#c) min cars sold per month
#d) average cars sold per month by gender
#e) average hours worked by people selling > 3 cars per month
#f) average years of experience
#g) average years of experience for people selling > 3 cars per month
#h) average cars sold per month by whether or not they have had sales training
#i) What do you think is the best indicator of whether someone is a good salesperson?
import pandas as pd
```

```python
Location = "datasets-1/axisdata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df.describe()
#a) average cars sold per month -3.92
#b) max cars sold per month - 7
#c) min cars sold per month - 1
#f) average years of experience - 3.02 years
```

```python
df_male=df[df['Gender']=='M']
df_male.describe()
```

```python
df_female=df[df['Gender']=='F']
df_female.describe()
```

```python
#d) average cars sold per month by gender: Male - 4.07 , Female - 3.76
```

```python
df_highsales=df[df["Cars Sold"] >3]
df_highsales.describe()
#e) average hours worked by people selling > 3 cars per month : 34.74hrs
#g) average years of experience for people selling > 3 cars per month : 3.04 years
```

```python
df_training=df[df['SalesTraining']=='Y']
df_training.describe()
```

```python
df_notraining=df[df['SalesTraining']=='N']
df_notraining.describe()
```

```python
#h) average cars sold per month by whether or not they have had sales training: had training - 4.16, no training - 3.57
```

```python
df_gender = pd.get_dummies(df['Gender'])
df_gender.tail()
```

```python
df_new = pd.concat([df, df_gender], axis=1)
df_new.tail()
```

```python
df_training = pd.get_dummies(df['SalesTraining'])
df_training.tail()
```

```python
df_new2 = pd.concat([df_new, df_training], axis=1)
df_new2.tail()
```

```python
df_new2.corr()
#i) What do you think is the best indicator of whether someone is a good salesperson? : Hours Worked has higher positive
# correlation with Cars Sold 0.25 so it is the best indicator of a good salesperson.
```
