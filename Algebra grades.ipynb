```python
#Algebra Grades Problem
#Frank Mulligrew is the Algebra Coordinator for DC Public Sc
#hools. He is required to gather some statistics by the
#School Board. Using the following information about his
#class, calculate the following:
#a) %age of students with a passing grade
#b) %age of women with a passing grade
#c) Average hours of study for all students
#d) Average hours of study for students with a passing grade
import pandas as pd
```

```python
Location = "datasets-1/algebradata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
#Convert gender in numbers: Female-1, Man- 0
def score_to_numeric(x):
    if x=='F':
        return 1
    if x=='M':
        return 0
```

```python
df['Sex_val'] = df['Gender'].apply(score_to_numeric)
df.tail()
```

```python
#Convert Grade in numbers: A-5, B-4, C-3, D-2, F-1
def score_to_numeric(y):
    if y=='A':
        return 5
    if y=='B':
        return 4
    if y=='C':
        return 3
    if y=='D':
        return 2
    if y=='F':
        return 1
```

```python
df['Grade_val'] = df['Grade'].apply(score_to_numeric)
df.tail()
df.describe()
#c) Average hours of study for all students - 15,115hrs
```

```python
df_passinggrade = df[df["Grade_val"] >=3]
df_passinggrade.head()
```

```python
df_passinggrade.describe()
#a) %age of students with a passing grade 691/999 = 69,17%
#d) Average hours of study for students with a passing grade - 16,298hrs
```

```python
df_femalepassinggrade = df_passinggrade[df_passinggrade["Sex_val"] ==1]
df_femalepassinggrade.head()
```

```python
df_femalepassinggrade.count()
#b) %age of women with a passing grade 381/999 = 38,14%
```
